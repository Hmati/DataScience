# Percentage of missing values

```python
housing.shape
```

```python
total_cells = np.product(housing.shape)

Missing_values = Missing_values.sum()

percentage_missing = (Missing_values/total_cells)*100

percentage_missing
```

The percentage of missing values is really small we can therefore drop it instead of doing Imputation on the data

```python
housing_copy=housing.copy()
housing_copy.head(2)
```

```python
housing_copy.shape
```

```python

housing_copy1=housing_copy.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

housing_copy1.shape
```

```python
housing_copy1.isna().sum()
housing_copy1.head()
```
